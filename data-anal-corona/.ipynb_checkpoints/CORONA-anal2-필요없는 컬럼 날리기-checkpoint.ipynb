```python
import pandas as pd
```

```python
# 데이터 파일 읽어오기 일별 서울확진자
df = pd.read_csv('corona/daySeoul.csv', encoding='UTF-8')
```

```python
df
```

```python
#필요없는 컬럼 날리기
df.drop(['released'], axis=1, inplace=True)
```

```python
#서울 일일사망자 추가
df['서울일일사망자']=df['death'].diff()
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df['서울일일사망자'] = df['서울일일사망자'].astype(int) 
```

```python
df
```

```python
df['사망률'] = (df['서울일일확진자'] / df['서울일일사망자']) * 100
```

```python
df
```

```python
df.drop(['사망률'], axis=1, inplace=True)
```

```python
df
```

```python
df['사망률'] = (df['confirmed'] / df['death']) * 100
```

```python
df
```

```python
df.fillna(0, inplace=True) 
```

```python
df
```

```python
df['사망률'] = df['사망률'].astype(int)
```

```python
import numpy as np
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
```

```python
df
```

```python
df.fillna(0, inplace=True) 
```

```python
df['사망률'] = df['사망률'].astype(int)
```

```python
df
```

```python
df.drop(['사망률'], axis=1, inplace=True)
```

```python
df
```

```python
df['사망률'] = (df['confirmed'] / df['death'])
```

```python
df
```

```python
df.drop(['사망률'], axis=1, inplace=True)
```

```python
df['사망률'] = (df['death'] / df['confirmed'])
```

```python
df
```

```python
df.drop(['사망률'], axis=1, inplace=True)
```

```python
df['사망률'] = (df['death'] / df['confirmed']) * 100
```

```python
df
```

```python
df.info()
```

```python
df['사망률']=df['사망률'].round(2)   
```

```python
df
```

```python
df.to_csv('CORONA/daySeoul.csv', sep=',')
```
