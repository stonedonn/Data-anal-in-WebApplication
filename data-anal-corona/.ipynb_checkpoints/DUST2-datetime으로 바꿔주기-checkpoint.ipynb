```python
import pandas as pd
# 데이터 파일 읽어오기 #12월
DUST = pd.read_csv('data/202012.csv', encoding='UTF-8', index_col=0)
```

```python
DUST.head(27)
```

```python
DUST.info()
```

```python
DUST['측정일시']=DUST['측정일시'].astype('str')
```

```python
# 측정일시 정규화하기
dust_date = []

for date in DUST['측정일시']:
        dust_date.append(str(date) + '0000')
```

```python
date=[]
for i in dust_date:
    date.append(str(i[:4]+'-'+i[4:6]+'-'+i[6:8]+' '+i[8:10]+':'+i[10:12]+':'+i[12:]))
    date
DUST['date']=pd.to_datetime(date)
date
#DUST.info()
```

```python
DUST.head(26)
```

```python
DUST.info()
```

```python
DUST.reset_index(drop='True')
```

```python
DUST.head(26)
```

```python
DUST_EX=DUST[['지역','망','측정소명','PM10','PM25','주소','date']]
```

```python
DUST = DUST_EX.reset_index(drop='True')
```

```python
DUST.head(27)
```
