```python
df = pd.read_csv('corona_Temperatures/corona_temperature.csv',encoding='UTF-8')
```

```python
df
```

```python
state_geo=json.load(open('corea.json', 'rt', encoding='UTF-8'))
```

```python
map=folium.Map(location=[35.8,127.6],tiles='OpenStreetMap',zoom_start=8)
folium.Choropleth(geo_data=state_geo,data=df,columns=['지역', '평균기온'],
                               key_on='feature.properties.id',fill_color='RdYlGn',legend_name='기온대비 확진자수 (%)').add_to(map)
for n in df.index:
    folium.CircleMarker([df['위도'][n], df['경도'][n]],
                        radius = df['누적확진자'][n]/1000,
                        color='#3186CC', fill_color='#3186CC', fill=True).add_to(map)
```
