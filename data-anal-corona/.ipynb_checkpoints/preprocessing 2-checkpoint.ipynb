```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./Data/201908.csv', encoding='utf-8',index_col=0)
```

```python
df.info()
df.head()
```

```python
df['측정일시']=df['측정일시'].astype(str)
df.info()
df=df.reset_index()
```

```python
datetime=[]
for i in df['측정일시']:
    datetime.append(str(i)+'0000')
datetime
```

```python
date=[]
for i in datetime:
    date.append(str(i[:4]+'-'+i[4:6]+'-'+i[6:8]+' '+i[8:10]+':'+i[10:12]+':'+i[12:]))
date
```

```python
df['date']=pd.to_datetime(date)
df.info()
```

```python
df.head()
```

```python
df=df.reset_index(drop=True)
df.head(25)
```

```python
df.drop('측정일시', axis=1, inplace=True)
df.head()
```

```python
df.drop('index',axis=1, inplace=True)
df.fillna(0,inplace=True)
df.head()
```

```python
dust_list=[]
for area in df['측정소명'].unique():
    dust_list.append(df[df['측정소명']==area].copy())
```

```python
for i, area in enumerate(df['측정소명'].unique()):
    dust_list[i].columns=['지역','망','측정소명','PM10','PM25','주소','date']
                          
```

```python
dust_list
```

```python
start='2019-08-01'
end='2019-08-31'
hourly_dust=pd.DataFrame(columns=['date'])
date_range=pd.date_range(start,end,freq='H')

hourly_dust['date']=date_range
```

```python
for d in dust_list:
    hourly_dust=hourly_dust.merge(d,how='outer')
```

```python
start='2019-08-01'
end='2019-08-31'
daily_dust=pd.DataFrame(columns=['date'])
date_range=pd.date_range(start,end,freq='D')
daily_dust['date']=date_range
```

```python
for d in range (daily_dust.shape[0]):
    for h in range (24):
        daily_dust.loc[d, 'dust_h'+str(h)]=hourly_dust['PM10'][d*24:d*24+24][d*24+h]
```

```python
daily_dust['avg']=daily_dust.loc[:,'dust_h0':].mean(axis=1)
daily_dust=daily_dust[['date','avg']]
daily_dust
```

```python
daily_dust.to_csv('./Data/PM10/Avg/201908.csv', sep=',')
```
