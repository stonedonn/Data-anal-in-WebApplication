```python
import pandas as pd
```

```python
# 데이터 파일 읽어오기 2021년 1월
df = pd.read_csv('dust/202101dust.csv', encoding='UTF-8')
```

```python
df.head()
```

```python
df['날짜'] = pd.to_datetime(df['날짜'])
```

```python
df.info()
```

```python
df['avg2.5'] = (df['서울']+df['부산']+df['대구']+df['인천']+df['광주']+df['대전']+df['울산']+df['경기']
             +df['강원']+df['충북']+df['충남']+df['전북']+df['전남']+df['경북']+df['경남']+df['제주']+df['세종']) / 17
```

```python
df.head()
```

```python
df=df.sort_values(by='날짜', ascending=True)
```

```python
df.head()
```

```python
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
df.to_csv('dust/202101dust.csv', sep=',')
```

```python
df1 = df[['날짜','avg2.5']]
```

```python
df1.head()
```

```python
df1.to_csv('dust/202101dustavg2.5.csv', sep=',')
```
