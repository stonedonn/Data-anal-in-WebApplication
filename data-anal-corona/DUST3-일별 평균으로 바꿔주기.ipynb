```python
import pandas as pd
# 데이터 파일 읽어오기 #5월
DUST = pd.read_csv('data/202005.csv', encoding='UTF-8', index_col=0)
```

```python
DUST.fillna(0, inplace=True)
```

```python
DUST.head(26)
```

```python
dust_list = []
```

```python
#지역별로 한 뭉텅이를 만들어서 하나의 리스트에 넣어준다.
for area in DUST['측정소명'].unique():
    dust_list.append(DUST[DUST['측정소명']==area].copy())
```

```python
dust_list
```

```python
#불필요해진 area열을 버리고, 대신 PM10, PM25 열에 area정보를 추가한다.
for i, area in enumerate(DUST['측정소명'].unique()):
    dust_list[i].columns=['지역','망','측정소명','PM10','PM25','주소','date']
                          
    
```

```python
# 새로운 데이터프레임 생성
start='2019-08-01'
end='2019-08-31'
hourly_dust=pd.DataFrame(columns=['date'])
date_range=pd.date_range(start,end,freq='H')

hourly_dust['date']=date_range
#hourly_dust.info()
date_range.info()
```

```python
for d in dust_list:
    hourly_dust=hourly_dust.merge(d,how='outer')
```
