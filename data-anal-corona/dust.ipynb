```python
# 데이터 파일 읽어오기 #12월
DUST = pd.read_csv('data/202012.csv', encoding='UTF-8', index_col=0)
```

```python
DUST.head()
```

```python
DUST.head()
```

```python
DUST.head(26)
```

```python
DUST['측정일시']=DUST['측정일시'].astype('str')
```

```python
# 측정일시 정규화하기
dust_date = []

for date in DUST['측정일시']:
        dust_date.append(str(date) + '0000')
```

```python
dust_date
```

```python
date=[]
for i in dust_date:
    date.append(str(i[:4]+'-'+i[4:6]+'-'+i[6:8]+' '+i[8:10]+':'+i[10:12]+':'+i[12:]))
    date
DUST['date']=pd.to_datetime(date)
date
#DUST.info()
```
